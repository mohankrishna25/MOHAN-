```python
def find_largest_simple(a, b, c):
    """Returns the largest of three numbers using built-in max()."""
    return max(a, b, c)

# Example usage:
num1 = 15
num2 = 42
num3 = 8

largest = find_largest_simple(num1, num2, num3)
print(f"The largest number is: {largest}")
```

```python
import calendar

# Example:
year_to_check = 2026
if calendar.isleap(year_to_check):
    print(f"{year_to_check} is a leap year!")
else:
    print(f"{year_to_check} is not a leap year.")
```

```python
def check_character(char):

    # Convert the character to lowercase for easier comparison
    char = char.lower()
    vowels = "aeiou"


    if char in vowels:
        return "Vowel"

    elif char.isalpha():
        return "Consonant"

    else:
        return "Neither (Not an alphabet letter)"

# Example usage:
print(check_character('A'))
print(check_character('z'))
print(check_character('7'))
```

```python
def check_divisibility_shortcut(number):

    return number % 55 == 0

# Example usage:
print(check_divisibility_shortcut(110))
print(check_divisibility_shortcut(60))
```

```python
def sum_natural_numbers():



    try:
        n = int(input("Enter a positive integer (N): "))


        if n <= 0:
            print("Please enter a positive integer greater than 0.")
            return


        total_sum = 0
        current_number = 1


        while current_number <= n:
            total_sum += current_number
            current_number += 1

        print(f"The sum of the first {n} natural numbers is: {total_sum}")

    except ValueError:
        print("Invalid input. Please enter a valid whole number.")


sum_natural_numbers()
```

```python
def print_multiplication_table():

    try:

        num = int(input("Enter a number to see its multiplication table: "))

        print(f"\n--- Multiplication Table for {num} ---")


        multiplier = 1


        while multiplier <= 15:


            result = num * multiplier


            print(f"{num} x {multiplier} = {result}")


            multiplier += 1

    except ValueError:
        print("Invalid input. Please enter a valid whole number.")

print_multiplication_table()
```
